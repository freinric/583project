```r
library(ggplot2)
library(dplyr)

myday = read.csv("Bike-Sharing-Dataset//day.csv", header = TRUE)
```

```r
ggplot(myday, aes(x = cnt, y = hum)) + 
  geom_point(alpha = 0.7) + 
  geom_smooth(method = "lm", se = FALSE) + 
  labs(title = "Scatterplot of Total Rental Bikes and Humidity", 
       x = "Total Rental Bikes (count)", y = "Humidity") 


ggplot(myday, aes(x = cnt, y = temp)) + 
  geom_point(alpha = 0.7) + 
  geom_smooth(method = "lm", se = FALSE) + 
  labs(title = "Scatterplot of Total Rental Bikes and Temperature", 
       x = "Total Rental Bikes (count)", y = "Humidity") 


ggplot(myday, aes(x = cnt, y = windspeed)) + 
  geom_point(alpha = 0.7) + 
  geom_smooth(method = "lm", se = FALSE) + 
  labs(title = "Scatterplot of Total Rental Bikes and Windspeed", 
       x = "Total Rental Bikes (count)", y = "Humidity") 
```
